# Preprocessing Data

```python
# Import necessary librarys for preprocessing

from sklearn.preprocessing import StandardScaler, FunctionTransformer, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics.pairwise import rbf_kernel
```

create a pipeline for the heavy_tail data

```python
heavy_tail_pipe = Pipeline(
    [
        ("log", FunctionTransformer(np.log, feature_names_out="one-to-one")),
        ("scale", StandardScaler()),
    ]
)
heavy_tail_pipe
```

create a new column for the shape_ratio

```python
# Divide the data into the train and the test
target = dataset["Class"]
dataset = dataset.drop("Class", axis=1)
target.shape, dataset.shape
```

split dataset into train and test

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    dataset, target, test_size=0.2, random_state=42, shuffle=True
)
```

```python
# Checking that in each round of test data and train classes are distributed fairly

print(f"Xtrain no Classes:{np.count_nonzero(np.unique(xtrain['class_int']))}")
print(f"Xtest no Classes:{np.count_nonzero(np.unique(xtest['class_int']))}")
```

drop the class_int column

```python
xtrain = xtrain.drop("class_int", axis=1)
xtest = xtest.drop("class_int", axis=1)
```

```python
# put all processing factors into a variable named all_transform

all_transform = ColumnTransformer(
    [
        (
            "heavy",
            heavy_tail_pipe,
            [
                "Area",
                "Perimeter",
                "MajorAxisLength",
                "MinorAxisLength",
                "ConvexArea",
                "EquivDiameter",
            ],
        ),
    ],
    remainder="passthrough",
)
all_transform
```

```python
# do all transform with train_transform command
train_transform = all_transform.fit_transform(xtrain)
test_transform = all_transform.transform(xtest)
```

```python
all_transform.get_feature_names_out()
```

plot data after transformation

```python
sns.set_style("white")
all_df = pd.DataFrame(train_transform, columns=all_transform.get_feature_names_out())
all_df.hist(bins=30, figsize=(17, 12))
plt.show()
```

# Import ML models from sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Import GridSearchCV for find best hyperparameters and models
from sklearn.model_selection import GridSearchCV, cross_val_score


from sklearn.metrics import (
    ConfusionMatrixDisplay,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
)

from sklearn.preprocessing import OneHotEncoder
```

# Find the best model using the GridSearchCV command

we use the GridSearchCV and all parameters for the best model

```python
pipe = Pipeline([("classifire", None)])

param_grids = [
    {
        "classifire": [SVC()],
    },
    {
        "classifire": [KNeighborsClassifier()],
        "classifire__n_neighbors": [1, 2, 3, 4, 5, 6],
        "classifire__n_jobs": [5, 10, 15],
    },
    {
        "classifire": [RandomForestClassifier()],
        "classifire__n_estimators": [300, 400],
        "classifire__max_depth": [20, 25],
        "classifire__n_jobs": [3, 5, 7],
    },
]
```

```python
gs = GridSearchCV(
    pipe,
    param_grid=param_grids,
    cv=5,
    verbose=1,
    scoring={"accuracy": "accuracy"},
    refit="accuracy",
)
gs
```

```python
# fit the model
gs.fit(train_transform, ytrain)
```

```python
# best estimator
gs.best_estimator_
model = gs.best_estimator_
model
```

```python
# best parameters for the best estimator
gs.best_params_
```

```python
# best score for the best estimator
gs.best_score_
```

```python
# predict model with test data
gs_pred = model.predict(test_transform)
```

```python
# model cross val scoring after 50 cv
gs_scores = cross_val_score(model, train_transform, ytrain, cv=50, scoring="accuracy")
```
