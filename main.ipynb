Explanation:

In this code snippet, we create a radar plot to visualize the values of multiple features for a specific sample. Radar plots, also known as spider or star plots, are useful for comparing the values of multiple variables across different categories or in this case, features. Each feature is represented by a spoke emanating from the center of the plot, and the value of the feature determines the distance of the point along the corresponding spoke.

Analysis:

By visualizing the radar plot of the sample features, we can analyze the relative magnitudes of each feature for the given sample. The distance of each point from the center of the plot represents the magnitude of the corresponding feature. We can compare the values of different features and observe any patterns or discrepancies among them. This visualization helps in understanding the distribution of feature values and identifying any outliers or unusual patterns in the data. Additionally, it provides insights into the relative importance or contribution of each feature to the overall sample. Overall, the radar plot aids in exploring the multivariate relationships among the features and gaining a holistic understanding of the sample characteristics.

```python
features = ["Area", "Perimeter", "MajorAxisLength", "MinorAxisLength"]


sample_index = 0
values = dataset.loc[sample_index, features].values.reshape(1, -1)


angles = np.linspace(0, 2 * np.pi, len(features), endpoint=False).tolist()


values = np.concatenate((values, values[:, [0]]), axis=1)
angles += angles[:1]


fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(polar=True))
ax.fill(angles, values.flatten(), color="red", alpha=0.25)
ax.plot(angles, values.flatten(), color="red", linewidth=2)
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(features)
plt.title("Radar Plot of Sample Features")
plt.show()
```

# Preprocessing Data

```python
# Import necessary librarys for preprocessing

from sklearn.preprocessing import StandardScaler, FunctionTransformer, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics.pairwise import rbf_kernel
```

create a pipeline for the heavy_tail data

```python
heavy_tail_pipe = Pipeline(
    [
        ("log", FunctionTransformer(np.log, feature_names_out="one-to-one")),
        ("scale", StandardScaler()),
    ]
)
heavy_tail_pipe
```

create a new column for the shape_ratio

```python
# Divide the data into the train and the test
target = dataset["Class"]
dataset = dataset.drop("Class", axis=1)
target.shape, dataset.shape
```

split dataset into train and test

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    dataset, target, test_size=0.2, random_state=42, shuffle=True
)
```

```python
# Checking that in each round of test data and train classes are distributed fairly

print(f"Xtrain no Classes:{np.count_nonzero(np.unique(xtrain['class_int']))}")
print(f"Xtest no Classes:{np.count_nonzero(np.unique(xtest['class_int']))}")
```

drop the class_int column

```python
xtrain = xtrain.drop("class_int", axis=1)
xtest = xtest.drop("class_int", axis=1)
```

```python
# put all processing factors into a variable named all_transform

all_transform = ColumnTransformer(
    [
        (
            "heavy",
            heavy_tail_pipe,
            [
                "Area",
                "Perimeter",
                "MajorAxisLength",
                "MinorAxisLength",
                "ConvexArea",
                "EquivDiameter",
            ],
        ),
    ],
    remainder="passthrough",
)
all_transform
```

```python
# do all transform with train_transform command
train_transform = all_transform.fit_transform(xtrain)
test_transform = all_transform.transform(xtest)
```

```python
all_transform.get_feature_names_out()
```

plot data after transformation

```python
sns.set_style("white")
all_df = pd.DataFrame(train_transform, columns=all_transform.get_feature_names_out())
all_df.hist(bins=30, figsize=(17, 12))
plt.show()
```

# Import ML models from sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Import GridSearchCV for find best hyperparameters and models
from sklearn.model_selection import GridSearchCV, cross_val_score


from sklearn.metrics import (
    ConfusionMatrixDisplay,
    confusion_matrix,
    classification_report,
    roc_curve,
    auc,
)

from sklearn.preprocessing import OneHotEncoder
```

# Find the best model using the GridSearchCV command

we use the GridSearchCV and all parameters for the best model

```python
pipe = Pipeline([("classifire", None)])

param_grids = [
    {
        "classifire": [SVC()],
    },
    {
        "classifire": [KNeighborsClassifier()],
        "classifire__n_neighbors": [1, 2, 3, 4, 5, 6],
        "classifire__n_jobs": [5, 10, 15],
    },
    {
        "classifire": [RandomForestClassifier()],
        "classifire__n_estimators": [300, 400],
        "classifire__max_depth": [20, 25],
        "classifire__n_jobs": [3, 5, 7],
    },
]
```

```python
gs = GridSearchCV(
    pipe,
    param_grid=param_grids,
    cv=5,
    verbose=1,
    scoring={"accuracy": "accuracy"},
    refit="accuracy",
)
gs
```

```python
# fit the model
gs.fit(train_transform, ytrain)
```

```python
# best estimator
gs.best_estimator_
model = gs.best_estimator_
model
```

```python
# best parameters for the best estimator
gs.best_params_
```

```python
# best score for the best estimator
gs.best_score_
```

```python
# predict model with test data
gs_pred = model.predict(test_transform)
```

```python
# model cross val scoring after 50 cv
gs_scores = cross_val_score(model, train_transform, ytrain, cv=50, scoring="accuracy")
```

 use classification_report to evaluate the model

```python
# model classification_report
print(classification_report(ytest, gs_pred))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

yy_train = le.fit_transform(ytest)
```

```python
import warnings


warnings.filterwarnings("ignore", category=UserWarning)


gs_prob = model.predict_proba(xtest)


fpr = {}
tpr = {}
roc_auc = {}
num_classes = 7

y_test_2d = np.zeros((len(yy_train), num_classes))
y_test_2d[np.arange(len(yy_train)), yy_train] = 1

for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_2d[:, i], gs_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])


all_fpr = np.unique(np.concatenate([fpr[i] for i in range(num_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(num_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= num_classes
roc_auc["macro"] = auc(all_fpr, mean_tpr)


plt.figure(figsize=(11, 7))
for i in range(num_classes):
    plt.plot(fpr[i], tpr[i], label=f"Class {i} (AUC = {roc_auc[i]:.2f})")


plt.plot([0, 1], [0, 1], linestyle="--", color="r", label="Random")


plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

Here we can see that the Sira class has the lowest accuracy and the highest error among other classes, and this can be due to its similarity with other classes close to it.
